```python
import pandas as pd
```

```python
import klib
```

```python
xlsx = pd.ExcelFile('sample_-_superstore.xls')
```

```python
df = pd.read_excel(xlsx, "Orders")
```

```python
df.info()
```

```python
klib.missingval_plot(df)
```

```python
df.head()
```

```python
df_cleaned = klib.data_cleaning(df)
```

```python
df_cleaned.info()
```

```python
df.info(memory_usage='deep')
```

```python
df_cleaned.info(memory_usage='deep')
```

```python
df_cleaned.head()
```

```python
df_profit = df_cleaned[["state", "profit"]].copy()
```

```python
df_profit.shape
```

```python
df_profit.info()
```

```python
df_profit.head(50)
```

```python
df_total_profit_by_state = df_profit.groupby("state").sum().reset_index()
```

```python
df_total_profit_by_state.head(50)
```

```python
df_total_profit_by_state ["rank"]= df_total_profit_by_state["profit"].rank(ascending=False)
```

```python
df_total_profit_by_state.shape
```

```python
df_total_profit_by_state.head(50)
```

```python
df_sorted = df_total_profit_by_state.sort_values("rank", ascending=True)
```

```python
df_sorted.shape
```

```python
df_sorted.head(50)
```

```python
df_sorted.info()
```

```python
df_roundup_profit = df_sorted.copy()
```

```python
df_roundup_profit.head(50)
```

```python
import numpy as np
```

```python
df_roundup_profit['profit'] = df_roundup_profit['profit'].apply(np.ceil).astype(int)
```

```python
df_roundup_profit.head(50)
```

```python
df_top_ten_states_by_profits = df_roundup_profit.query("rank<=10") 
```

```python
df_top_ten_states_by_profits.head(11)
```

```python
df_top_ten_states_by_profits["profit"].describe()
```

```python
df_top_ten_states_by_profits.to_csv('topTenStatesByProfits.csv', index=False)
```
